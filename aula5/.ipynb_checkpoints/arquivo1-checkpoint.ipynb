```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
imagem = cv2.imread('ju.jpg')
```

```python
type(imagem)
```

```python
plt.imshow(imagem)
```

```python
imagem_nova = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem_nova)
```

```python
imagem_nova.shape
```

```python
mud_img = cv2.resize(imagem_nova, (2080, 1560))
```

```python
plt.imshow(mud_img)
```

```python
mud_img.shape
```

```python
lag = 0.2
alt = 0.2
```

```python
outra_imagem = cv2.resize(imagem_nova, (0,0) , imagem_nova, lag, alt)
```

```python
plt.imshow(outra_imagem)
```

```python
imagem_gir = cv2.flip(imagem_nova, -1)
```

```python
plt.imshow(imagem_gir)
```

```python
cv2.imwrite('ju.jpg', imagem)
```

```python
img = plt.figure(figsize=(10,8))
ax = img.add_subplot(111)
ax.imshow(imagem_gir)
```
