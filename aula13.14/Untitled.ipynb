```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xadrez = cv2.imread('xadrez.jpeg')
xadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2RGB)
cinzax = cv2.cvtColor(xadrez, cv2.COLOR_RGB2GRAY)
plt.imshow(xadrez)
```

```python
bordas = cv2.goodFeaturesToTrack(cinzax, 50, 0.01, 10)
```

```python
bordas
```

```python
bordas = np.int0(bordas)
```

```python
bordas
```

```python
for i in bordas:
    x,y = i.ravel()
    cv2.circle(xadrez, (x,y), 3, (0,255,0), -1)

plt.imshow(xadrez)
```
